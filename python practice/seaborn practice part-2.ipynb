### Bar Plot

these plots allows you to get a aggregrate data off a categorical feature in your data.
barplot is a general plot that allows you to aggregate the categorical data based of the some function.


countplot : Show the counts of observations in each categorical bin.    
pointplot : Show point estimates and confidence intervals using scatterplot
            glyphs.    
catplot : Combine a categorical plot with a class:`FacetGrid`. 

```python
sns.barplot(x='sex',y='total_bill',data=tips)#by default the estimator function is mean
```

```python
#changing the estimator function to standard deviation.
sns.barplot(x='sex',y='total_bill',data =tips, estimator=np.std)
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
from numpy import median
sns.barplot(x="day", y="tip", data=tips, estimator=median)
```

```python
sns.barplot(x="day", y="tip", data=tips, ci=68)
```

```python
sns.barplot(x="day", y="tip", data=tips, ci="sd")
```

```python
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
tips.head()
```

```python
sns.barplot(x="day", y="total_bill", hue="weekend",data=tips, dodge=False,palette='coolwarm')
```

```python
sns.barplot("size", y="total_bill", data=tips,color="salmon", saturation=.5)#single color
```

catplot:-to combine a :func:`barplot` and a :class:`FacetGrid`.
This allows grouping within additional categorical variables. Using
:func:`catplot` is safer than using :class:`FacetGrid` directly, as it
ensures synchronization of variable order across facets:

```python
sns.catplot(x="sex", y="total_bill",hue="smoker", col="time",data=tips, kind="bar",height=4, aspect=.7);
```

```python
sns.barplot(x="day", y="tip", data=tips, capsize=.2)
```

```python
sns.barplot("size", y="total_bill", data=tips,palette="Reds")
```

```python
#Draw a set of vertical bars with nested grouping by a two variables:
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)#using a hue parameter.
```

```python
sns.barplot(x="tip", y="day", data=tips,hue='sex')
```

```python
sns.barplot(x="time", y="tip", data=tips,order=["Lunch", "Dinner"])
```

```python
sns.barplot("day", "total_bill", data=tips,linewidth=2.5, facecolor=(1, 1, 1, 0),errcolor=".2", edgecolor=".2")
```

### Box Plot

these types of plots are used to show that distribution of categorical data, a box plot is also a sometimes known as whisker plots and its shows the distribution of quantitative data in a way that hopefully facilitates comparisons variable.

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
#Draw a boxplot with nested grouping by two categorical variables:
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="Set3")
```

```python
#Draw a boxplot with nested grouping when some bins are empty:
sns.boxplot(x="day", y="total_bill", hue="time",data=tips, linewidth=2.5)
```

```python
sns.boxplot(x="time", y="tip", data=tips,order=["Dinner", "Lunch"])
```

```python
iris = sns.load_dataset("iris")
ax = sns.boxplot(data=iris, orient="h", palette="Set2")#using a orient parameter
```

```python
sns.boxplot(x="day", y="total_bill", hue="weekend",data=tips, dodge=False)
```

### Violin Plot
these types of plots are used to show that distribution of categorical data.
A violin plot plays a similar role as a box and whisker plot. It shows the
distribution of quantitative data across several levels of one (or more)
categorical variables such that those distributions can be compared. Unlike
a box plot, in which all of the plot components correspond to actual
datapoints, the violin plot features a kernel density estimation of the
underlying distribution.

This can be an effective and attractive way to show multiple distributions
of data at once, but keep in mind that the estimation procedure is
influenced by the sample size, and violins for relatively small samples
might look misleadingly smooth.

```python
sns.violinplot(x='day',y='total_bill', data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.violinplot(x="day", y="total_bill", hue="smoker",data=tips, palette="muted")
```

```python
sns.violinplot(x="time", y="tip", data=tips,order=["Dinner", "Lunch"])
```

```python
#Scale the violin width by the number of observations in each bin:
sns.violinplot(x="day", y="total_bill", hue="sex",data=tips, palette="Set2", split=True,scale="count")
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex",data=tips, palette="Set2", split=True,scale="count", inner="quartile")
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex",data=tips, palette="Set2", split=True,scale="count", inner="stick")
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex",data=tips, palette="Set2", split=True,scale="count", inner="stick", scale_hue=False)
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex",data=tips, palette="Set2", split=True,scale="count", inner="stick",scale_hue=False, bw=.2)
```

```python
planets = sns.load_dataset("planets")
sns.violinplot(x="orbital_period", y="method",data=planets[planets.orbital_period < 1000],scale="width", palette="Set3")
```

```python
sns.violinplot(x="orbital_period", y="method",data=planets[planets.orbital_period < 1000],cut=0, scale="width", palette="Set3")
```

```python
sns.violinplot(x="day", y="total_bill", hue="weekend", data=tips, dodge=False)
```

### Strip plot

It is going to a plot of a scatter plot in which one variable is categorical.

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
#adding hue and strip arguemnt
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True)
```

### swarm Plots
it is the combination of voilin plot ad strip plot. It is very much similiar to strip plot except the points are adjusted so that they don't overlap so we get a better plot of a distribution value.

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
#combining voilin plot and swarm plot
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

### Factor Plot

This function provides access to several axes-level functions that
show the relationship between a numerical and one or more categorical
variables using one of several visual representations. The ``kind``
parameter selects the underlying axes-level function to use:


```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```
